```python
import numpy as np
from matplotlib import pyplot as plt
```

### Definición de la función _simular_PAS_

```python
def simular_PAS(p,n,reps):
  # Input:
  # p: valor entre 0 y 1 que indica la probabilidad de tomar el valor 1 en cada paso (prob 1-p para tomar el -1)
  # n: número (natural) de pasos que vamos a simular en cada trayectoria
  # reps: cantidad de trayectorias que vamos a representar.

  # Generamos la matriz de pasos (X_i) de tamaño reps*n
  Matriz_pasos = (np.random.rand(reps, n) < p) * 2 - 1

  # Calcular la suma acumulada por columnas (axis=1)
  # Esto implementa la matriz en la que la columna t es X_t = Z_1 + Z_2 + ... + Z_t para todas las reps
  matriz_sumas= np.cumsum(Matriz_pasos, axis=1)

  return matriz_sumas
```

### Explicación de la función:
  La sentencia *np.random.rand(reps, n)* nos genera una matriz de dimensión reps x n con números entre 0 y 1 aleatoriamente.

  La matriz pasará a ser una matriz _booleana_ (solo con 0 y 1) tras realizar la comprobación de que cada elemento sea mayor o menor que p, en cuyo caso, tendremos un 0 o un 1, respectivamente.

  El multiplicar *2 y restar -1 es una forma de que nuestra matriz solo tenga 1 y -1, pues:
  - Si el elemento es $0 \implies (0\times2) - 1 = -1$
  - Si el elemento es $1 \implies (1\times2) - 1 = 1$


## Visualización 1: diferentes trayectorias del PAS (sin media ni desviación típica teóricas)

A continuación se presenta el código para crear las imágenes de las diapositivas 16 ($p=0.5$), 17 ($p=0.6$) y 18 ($p=0.4$) del tema 1.

```python
p1=0.5
reps=5
n=500

#Definimos una variables donde almacenar la matriz que nos devuelve la función
output1=simular_PAS(p1,n,reps)

#Graficar el PAS
plt.plot(output1.T)
plt.show()
```

## Visualización 2: diferentes trayectorias del PAS (con media y desviación típica teóricas)

Ahora, el código para crear las imágenes de las diapositivas 30 ($p=0.5$), 31 ($p=0.75$), 32 ($p=0.95$) del tema 1.

```python
p1=0.5
reps=20
n=500

#Guardar el resultado de la función
output1=simular_PAS(p1,n,reps)

#Crear un array para el eje horizontal (de 0 a n) para después calcular la media y la varianza
t=np.arange(n)

#Calcular la media y varianza
media_teorica1=t*(2*p1-1)
sigma1=np.sqrt(t*4*p1*(1 - p1))

# Primero dibujamos las bandas para que queden por debajo de las líneas
# Entorno +- 2 sigma
plt.fill_between(t, media_teorica1 - 2*sigma1, media_teorica1 + 2*sigma1, color='silver', label='$\\pm 2\\sigma$')

# Entorno +- 1 sigma
plt.fill_between(t, media_teorica1 - sigma1, media_teorica1 + sigma1, color='gray', label='$\\pm 1\\sigma$')

#Crear el PAS
plt.plot(t, output1.T, linewidth=0.6, alpha=0.9)

# Dibujar la línea de la media teórica
plt.plot(t, media_teorica1, color='black', linestyle='--', linewidth=2, label='Media teórica')

# Añadir etiquetas a los ejes
plt.title('20 trayectorias de un PAS con p=0,5')
plt.xlabel('$t$')
plt.ylabel('$X_t$')

# Configurar la leyenda
plt.legend(loc='upper left')

plt.show()
```

```python
p2=0.75
reps=20
n=500

#Guardar el resultado de la función
output2=simular_PAS(p2,n,reps)

#Crear un array para el eje horizontal (de 0 a n) para después calcular la media y la varianza
t = np.arange(n)

#Calcular la media y varianza
media_teorica2 = t*(2*p2-1)
sigma2=np.sqrt(t*4*p2*(1 - p2))

# Primero dibujamos las bandas (para que queden por debajo de las líneas)
# Entorno +- 2 sigma
plt.fill_between(t, media_teorica2 - 2*sigma2, media_teorica2 + 2*sigma2, color='silver', label='$\\pm 2\\sigma$')

# Entorno +- 1 sigma
plt.fill_between(t, media_teorica2 - sigma2, media_teorica2 + sigma2, color='gray', label='$\\pm 1\\sigma$')

#Crear el PAS
plt.plot(t, output2.T, linewidth=0.6, alpha=0.9)

# Dibujar la línea de la media teórica
plt.plot(t, media_teorica2, color='black', linestyle='--', linewidth=2, label='Media teórica')

# Añadir etiquetas a los ejes
plt.title('20 trayectorias de un PAS con p=0,75')
plt.xlabel('$t$')
plt.ylabel('$X_t$')

# Configurar la leyenda
plt.legend(loc='upper left')

plt.show()
```
